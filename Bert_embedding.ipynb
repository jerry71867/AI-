```python
from transformers import AutoTokenizer, AutoModel
#匯入AutoTokenizer和AutoModel類別，提供預訓練的語言模型和相關工具。
model=AutoModel.from_pretrained("bert-base-uncased");
#使用AutoModel的from_pretrained方法來載入一個預訓練的BERT模型。
print(model)
#顯示有關模型的詳細信息。
```

```python
print(model.num_parameters())
#打印出載入的模型中的參數數量。
```

```python
from transformers import AutoTokenizer
#從transformers庫中匯入AutoTokenizer類別，提供預訓練的語言模型和相關工具。
tokenizer=AutoTokenizer.from_pretrained("bert-base-uncased")
#創建一個AutoTokenizer的實例，並從預訓練模型中載入對應的tokenizer，將文本轉換成模型可以理解的數字表示形式。
raw_inputs = [
    "I've been waiting for a HuggingFace course my whole life.",
    "I hate this so much!",
]
#定義了一個包含兩個字符串的列表raw_inputs。
inputs = tokenizer(raw_inputs, padding=True, truncation=True, return_tensors="pt")
"""
使用tokenizer將原始文本轉換為模型所需的輸入格式。
padding=True表示對輸入進行填充，
truncation=True表示對長度進行截斷，
return_tensors="pt"表示返回PyTorch張量。
"""
print(inputs)
#打印轉換後的輸入，以便查看其格式和結構。
```
