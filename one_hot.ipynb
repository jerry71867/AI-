```python
import numpy as np
#匯入numpy模組，用於處理數值數據和多維數組。
import pandas as pd
#匯入pandas模組，用於數據操作和分析。
country=['Taiwan','Australia','Ireland','Australia','Ireland','Taiwan']
#定義了一個包含國家名稱的列表country。
age=[25,30,45,35,22,36]
#定義了一個包含年齡的列表age。
salary=[20000,32000,59000,60000,43000,52000]
#定義了一個包含薪資的列表 salary。
dic={'Country':country,'Age':age,'Salary':salary}
#定義了一個字典dic。
data=pd.DataFrame(dic)
#使用pandas的DataFrame函數，將字典dic轉換為一個數據框data。
data
#用來顯示data數據框的內容
```

```python
from sklearn.preprocessing import LabelEncoder
#從sklearn.preprocessing模組中匯入LabelEncoder類別，用於將分類數據（如字符串標籤）轉換為數字編碼。
labelencoder = LabelEncoder()
#創建一個LabelEncoder類別的實例，並將其賦值給變數labelencoder。
data_le=pd.DataFrame(dic)
#創建一個pandas數據框data_le，和data數據框相同，包含'Country'、'Age'和'Salary'三列。
data_le['Country'] = labelencoder.fit_transform(data['Country'])
#對data數據框中的'Country'列應用標籤編碼。
data_le
#顯示data_le數據框的內容。
```

```python
from sklearn.preprocessing import OneHotEncoder
#從sklearn.preprocessing模組中匯入OneHotEncoder類別，用於將分類數據轉換為一個熱編碼表示。
from sklearn.compose import ColumnTransformer
#從sklearn.compose模組中匯入ColumnTransformer類別，用於對特定列應用不同的轉換器
ct = ColumnTransformer([("Country", OneHotEncoder(), [0])], remainder = 'passthrough')
"""
創建了一個 ColumnTransformer 實例
ColumnTransformer指定對索引為0的列（即'Country'列）應用OneHotEncoder，而對其餘列則保持不變（passthrough）。
"""
data_str_ohe=ct.fit_transform(data_le)
#對數據框data_le應用ColumnTransformer。
pd.DataFrame(data_str_ohe)
#將data_str_ohe數組轉換為一個pandas數據框
```

```python

data_dum = pd.get_dummies(data)
#使用pd.get_dummies()函數對數據框data中的每個分類特徵進行一個熱編碼（one-hot encoding）轉換。
pd.DataFrame(data_dum)
#將data_dum數據框轉換為一個新的pandas數據框。
```
