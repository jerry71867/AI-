```python
from transformers import pipeline
#匯入pipeline函式，用於輕鬆地使用預訓練模型進行各種NLP任務。

classifier = pipeline("sentiment-analysis")
#下載並使用預訓練的情感分析模型。
classifier("你敢出去玩試看看")
#返回一個包含情感分析結果的字典，包括預測的情感類別和相應的置信度分數。
```

```python
classifier(
    ["我覺得東西不好吃", "I love this cake"]
)
#返回一個包含兩個文本情感分析結果的列表，包括預測的情感類別和相應的置信度分數。
```

```python
from transformers import pipeline
#匯入pipeline函式，用於輕鬆地使用預訓練模型進行各種NLP任務。

classifier = pipeline("zero-shot-classification")
#自動下載並使用預訓練的零標籤分類模型。
classifier(
    "你敢出去玩試看看",
    candidate_labels=["出門", "留在家", "再問一次"],
)
#返回一個字典，包含預測的標籤以及相應的置信度分數。
```

```python
from transformers import pipeline
#匯入pipeline函式，用於輕鬆地使用預訓練模型進行各種NLP任務。

generator = pipeline("text-generation")
#自動下載並使用預訓練的文本生成模型。
generator("我等一下要去", num_return_sequences = 2, max_length = 30)
#返回一個包含生成文本序列的列表。
```

```python
generator("我等一下想吃", num_return_sequences = 2, max_length = 30)
#返回一個包含兩個生成文本序列的列表。
```

```python
from transformers import pipeline
#匯入pipeline函式，用於輕鬆地使用預訓練模型進行各種NLP任務。

generator = pipeline("text-generation", model="ckiplab/gpt2-base-chinese")
"""
參數"text-generation"指定了所需的NLP任務。
model="ckiplab/gpt2-base-chinese"指定了使用中文GPT-2模型進行文本生成。
"""
generator(
    "今天天氣真好，",
    max_length=30,
    num_return_sequences=2,
)
#返回一個包含兩個生成文本序列的列表。
```

```python
from transformers import pipeline
#匯入pipeline函式，用於輕鬆地使用預訓練模型進行各種NLP任務。

unmasker = pipeline("fill-mask")
#自動下載並使用預訓練的遮罩填充模型。
unmasker("這堂課正在教人工<mask>慧有關於大型語<mask>模型.", top_k=1)
```

```python
from transformers import pipeline
#匯入pipeline函式，用於輕鬆地使用預訓練模型進行各種NLP任務。

ner = pipeline("ner", grouped_entities=True, model="gyr66/RoBERTa-ext-large-chinese-finetuned-ner")
"""
參數"ner"指定了所需的NLP任務。
grouped_entities=True表示要將識別的命名實體分組，
model="gyr66/RoBERTa-ext-large-chinese-finetuned-ner"指定了使用RoBERTa預訓練模型進行命名實體識別。
"""
ner("小明去參加苗栗的白沙屯媽祖繞境")
#返回一個包含識別的命名實體及其類別的列表。
```

```python
from transformers import pipeline
#匯入pipeline函式，用於輕鬆地使用預訓練模型進行各種NLP任務。

ner = pipeline("ner", model = "ckiplab/bert-base-chinese-ner", grouped_entities=True)
"""
參數"ner"指定了所需的NLP任務。
model = "ckiplab/bert-base-chinese-ner"指定了使用CKIP實驗室的中文BERT預訓練模型進行命名實體識別。
"""
ner("我的名字叫陳小明，我在新竹明新科大上課。")
#返回一個包含識別的命名實體及其類別的列表。
```

```python
from transformers import pipeline
#匯入pipeline函式，用於輕鬆地使用預訓練模型進行各種NLP任務。

question_answerer = pipeline("question-answering", model="ckiplab/bert-base-chinese-qa")
"""
參數"question-answering"指定了所需的NLP任務，即問答系統。
model="ckiplab/bert-base-chinese-qa"指定了使用CKIP實驗室的中文BERT預訓練模型進行問答。
"""
question_answerer(
    context="黃小明參加大甲鎮瀾宮繞境",
    question="黃小明參加甚麼活動?",
)
#返回一個包含問答結果的字典，包括提取的答案以及相應的分數。
```

```python
from transformers import pipeline
#匯入pipeline函式，用於輕鬆地使用預訓練模型進行各種NLP任務。

summarizer = pipeline("summarization")
#自動下載並使用預訓練的摘要生成模型。
summarizer(
    """
    台北市知名連鎖韓式炸雞店日前遭導盲犬使用者投訴，攜帶導盲犬前往用餐時，卻遭服務人員拒絕入內，事後將過程放上臉書，
    店家事後也發文道歉，北市社會局也說，以要求業者限期改善，對員工教育訓練，若未改善或再犯最高將開罰5萬元，
    並接受4小時講習
"""
)
#返回一個包含摘要的字串，包含了提取的關鍵信息。
```

```python
from transformers import pipeline
#匯入pipeline函式，用於輕鬆地使用預訓練模型進行各種NLP任務。

translator = pipeline("translation", model="DunnBC22/opus-mt-zh-en-Chinese_to_English")
"""
參數"translation"指定了所需的NLP任務，即翻譯。
model="DunnBC22/opus-mt-zh-en-Chinese_to_English"指定了使用特定模型進行中文到英文的翻譯。
"""
translator("我參加大甲鎮瀾宮pilgrimage.")
#返回一個包含翻譯結果的字串。
```
