```python
from datasets import load_dataset
#匯入load_dataset函式。用於加載指定的數據集。

raw_datasets = load_dataset("glue", "mrpc")
"""
加載GLUE基準數據集中的MRPC（Microsoft Research Paraphrase Corpus）數據集。
GLUE（General Language Understanding Evaluation）是一個自然語言理解的基準數據集集合
"""
raw_datasets
#返回並顯示加載的 MRPC 數據集。
```

```python
raw_train_dataset = raw_datasets["train"]
#從先前加載的raw_datasets中提取訓練數據集部分，並將其賦值給變數raw_train_dataset。
raw_train_dataset[0]
#從raw_train_dataset中提取第一個樣本。
```

```python
raw_train_dataset.features
"""
從raw_train_dataset提取features屬性。
features屬性包含數據集中每個字段的描述和類型信息。
"""
```

```python
from transformers import AutoTokenizer
#匯入AutoTokenizer，根據指定的預訓練模型名稱自動選擇並加載對應的分詞器。

checkpoint = "bert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(checkpoint)
tokenized_sentences_1 = tokenizer(raw_datasets["train"]["sentence1"])
tokenized_sentences_2 = tokenizer(raw_datasets["train"]["sentence2"])
```

```python
inputs = tokenizer("This is the first sentence.", "This is the second one.")
#使用先前加載的tokenizer分詞器對兩個句子進行編碼，將它們轉換為模型可以處理的輸入格式。
inputs
#返回並顯示分詞器處理後的結果。
```

```python

tokenizer.convert_ids_to_tokens(inputs["input_ids"])
#使用分詞器（tokenizer）將輸入中的token IDs轉換回對應的token字符串。
```

```python
tokenized_dataset = tokenizer(
    raw_datasets["train"]["sentence1"],
    raw_datasets["train"]["sentence2"],
    padding=True,
    truncation=True,
)
"""
將raw_datasets訓練集中的句子對進行編碼，
並生成一個包含token IDs、attention mask等輸入格式的字典，適合用於後續的模型訓練或推理。

"""
```

```python
def tokenize_function(example):
#定義了一個名為tokenize_function的函數。
    return tokenizer(example["sentence1"], example["sentence2"], truncation=True)
    #使用分詞器將example中的兩個句子進行編碼，並返回編碼結果。
```

```python
tokenized_datasets = raw_datasets.map(tokenize_function, batched=True)
"""
將tokenize_function函數應用於原始數據集中的每個樣本，
並以批量方式進行處理，生成一個新的數據集tokenized_datasets。
"""
tokenized_datasets
#顯示tokenized_datasets，包含原始數據集中所有樣本經過tokenize_function函數處理後的結果。
```

```python
from transformers import DataCollatorWithPadding
#導入DataCollatorWithPadding，用於處理訓練數據以準備訓練模型。

data_collator = DataCollatorWithPadding(tokenizer=tokenizer)
#使用分詞器tokenizer來處理輸入數據，並將其填充到相同的長度，以便用於模型的訓練。
```

```python
samples = tokenized_datasets["train"][:8]
#從tokenized_datasets中的訓練集中選擇了前8個樣本，將其存儲在samples變數中。
samples = {k: v for k, v in samples.items() if k not in ["idx", "sentence1", "sentence2"]}
#移除鍵為"idx"、"sentence1"和"sentence2"的項目，去除不必要的資訊，專注於分析和處理模型所需的輸入資料。
[len(x) for x in samples["input_ids"]]
#計算了每個樣本中input_ids列表中的元素個數（即每個樣本的序列長度），並返回一個列表。
```

```python
batch = data_collator(samples)
#使用之前定義的data_collator將提取的樣本samples整理成一個批次資料。
{k: v.shape for k, v in batch.items()}
#列印了批次資料batch中每個輸入張量的形狀。
```
